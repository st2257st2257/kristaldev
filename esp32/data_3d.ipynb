```python
import time
import requests
response = requests.get('http://kristaldev.online:8092/functions/sensors/get/2')
if response.status_code == 200:
    print (str(response.content).split(' '))
```

```python
res_data = []
n = 6000
for i in range(n):
    res = []
    data = str(requests.get('http://kristaldev.online:8092/functions/sensors/get/2').content).split(' ')
    res.append(data[1])
    res.append(data[2])
    res.append(data[8])
    res_data.append(res)
    time.sleep(0.1)
```

```python
# print(res_data)
```

```python
import matplotlib.pyplot as plt
import numpy as np


x_av = sum([float(item[0]) for item in res_data])/n
y_av = sum([float(item[1]) for item in res_data])/n

x = [(float(item[0])-x_av)*100000 for item in res_data]#np.linspace(-4*np.pi,4*np.pi,50)
y = [(float(item[1])-y_av)*100000 for item in res_data]#np.linspace(-4*np.pi,4*np.pi,50)
z = [float(item[2]) for item in res_data]#x**2 + y**2

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')

ax.plot(x,y,z)

plt.show()
```
